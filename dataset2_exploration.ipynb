# Data preprocessing

The package 'scanpy' was utilized to bring the data frames into an appropriate form to continue the scRNA seq pipeline. This package is the equivalent of the 'Seurat' package of R, that was used in the paper's analysis. 

Following the paper's instructions genes with expression in less than three cells, as well as cells with less than 900 genes, were removed. In the filtered data, FPKM values were transformed into log-space.

After that, PCA was performed using highly variable genes in the trimmed dataset. The first two PCâ€™s were used as input to perform density clustering in order to identify groupings in the data. T-distributed statistical neighbor embedding (tSNE) was utilized to visualize. Using further scanpy functionality, marker genes for each respective cluster were identified and used for subsequent analysis. 

```python
data_id4_copy = data_id4.copy()
data_id4_copy = data_id4_copy.T
```

```python
# create annotated data object (with metadata about cell type and gene ids)
adata_id4 = sc.AnnData(data_id4_copy)
adata_id4
```

```python
# transform data to log space
sc.pp.log1p(adata_id4)
adata_id4
```

```python
# filter genes based on the minimum expression threshold
sc.pp.filter_genes(adata_id4, min_cells=3)
sc.pp.filter_cells(adata_id4, min_genes=500)
sc.pp.filter_cells(adata_id4, max_genes=6000)
```

```python
# filter out high mitochondrial cell expression
mito_genes = adata_id4.var_names.str.startswith('MT-')
adata_id4.obs['percent_mito'] = np.sum(adata_id4[:, mito_genes].X, axis=1) / np.sum(adata_id4.X, axis=1)
adata_id4 = adata_id4[adata_id4.obs['percent_mito'] <= 0.1]
```

```python
# perform data normalization and scaling
sc.pp.normalize_total(adata_id4)
#sc.pp.log1p(adata_id4)
sc.pp.scale(adata_id4)
adata_id4
```

```python
adata_id4.obs
```

```python
adata_id4.var
```

```python
# perform regression to remove unwanted sources of variation
sc.pp.regress_out(adata_id4, keys=['n_genes', 'percent_mito'])
```

```python
# identify high variable genes for the PCA
sc.pp.highly_variable_genes(adata_id4, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
sc.pl.highly_variable_genes(adata_id4, show=False)
```

```python
# perform PCA using the identified highly variable genes
sc.tl.pca(adata_id4, svd_solver='arpack', use_highly_variable=True)
```

```python
# compute t-SNE using the first ten principal components
sc.tl.tsne(adata_id4, n_pcs=10)
```

```python
cell_ids = adata_id4.obs.index
cell_ids
```

```python
sc.pl.tsne(adata_id4)
```

```python
# clustering with leiden algorithm
sc.pp.neighbors(adata_id4, n_pcs=10)
sc.tl.leiden(adata_id4)
```

```python
# identify marker genes for each cluster
sc.tl.rank_genes_groups(
    adata_id4,
    groupby='leiden',
    method='t-test',
    use_raw=False,
    log_transform=False,
    min_in_group_fraction=0.01
)
```

```python
basal_marker_genes = ['KRT5', 'KRT14']
luminal_marker_genes = ['KRT8', 'KRT18', 'ANKRD30A']
all_marker_genes = basal_marker_genes + luminal_marker_genes

sc.pl.dotplot(adata_id4, all_marker_genes, groupby='leiden')
```
