```python
import pandas as pd
```

```python
#clustering_results = pd.read_csv('ind4_exp_matrix_with_clusters.csv.gz', compression='gzip')
clustering_results = pd.read_csv('ind4_var_genes_with_clusters.csv.gz', compression='gzip')
```

```python
clustering_results
```

```python
# Map the cluster labels to their corresponding names
label_mapping = {
    0: 'L1',
    1: 'L1',
    2: 'B',
    3: 'B',
    4: 'L2',
    5: 'B',
    6: 'L2',
    7: 'Unclassified',
    8: 'Unclassified'
}

# Replace the values in the 'cluster_label' column
clustering_results['cluster_label'] = clustering_results['cluster_label'].map(label_mapping)

clustering_results.head()
```

```python
print(clustering_results['cluster_label'].value_counts())
```
